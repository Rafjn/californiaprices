```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from sklearn.compose import ColumnTransformer
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import ElasticNet, LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import (
    OrdinalEncoder,
    OneHotEncoder,
    PolynomialFeatures,
    PowerTransformer,
    RobustScaler,
    StandardScaler,
    QuantileTransformer,
)

from src.config import DADOS_LIMPOS
from src.auxiliares import dataframe_coeficientes
from src.graficos import (
    plot_coeficientes,
    plot_comparar_metricas_modelos,
    plot_residuos_estimador,
)
from src.models import (
    organiza_resultados,
    treinar_e_validar_modelo_regressao,
    grid_search_cv_regressor,
)
```

```python
df = pd.read_parquet(DADOS_LIMPOS)

df.head()
```

```python
df.info()
```

```python
#Target
coluna_target = ['median_house_value'] 

#categorias que n√£o tem ordem
coluna_one_hot_encoder = ['ocean_proximity']

#categorias que tem ordem
coluna_ordinal_encoder = ['median_income_cat'] 

```

```python
#colunas robust scaler
colunas_robust_scaler = df.columns.difference(coluna_one_hot_encoder + coluna_ordinal_encoder + coluna_target)

#colunas standard scaler
colunas_standard_scaler = ['longitude', 'latitude', 'housing_median_age'] #Colunas que tem distribui√ß√£o parecida bimodal

#colunas power transform
colunas_power_scaler = df.columns.difference(coluna_one_hot_encoder + coluna_ordinal_encoder + coluna_target + colunas_standard_scaler)
```

```python
#criando meu preprocessamento de colunas


#Aplico apenas o preprocessamento das colunas categ√≥ricas
preprocessamento_categoricas = ColumnTransformer(
    transformers=[
        ('ordinal_encoder', OrdinalEncoder(categories='auto'), coluna_ordinal_encoder), #O auto auto reconhece as categorias
        ('one_hot_encoder', OneHotEncoder(drop='first'), coluna_one_hot_encoder)
    ],
    remainder='passthrough' #N√£o fa√ßa nada com as demais colunas
)

#Aplicando o robust scaler a todas as colunas numericas
preprocessamento_simples = ColumnTransformer(
    transformers=[
        ('ordinal_encoder', OrdinalEncoder(categories='auto'), coluna_ordinal_encoder), #O auto auto reconhece as categorias
        ('one_hot_encoder', OneHotEncoder(drop='first'), coluna_one_hot_encoder),
        ('robust_scaler', RobustScaler(), colunas_robust_scaler) #Parecido com o standard, mas menos sensivel a outliers(bom de usar quando minha base tem muitos outliers)
    ],
)


#Aplico o standard scaler nas bimodais e o power transform nas demais + preprocessamento de categ√≥ricas
preprocessamento_completo = ColumnTransformer(
    transformers=[
        ('ordinal_encoder', OrdinalEncoder(categories='auto'), coluna_ordinal_encoder), #O auto auto reconhece as categorias
        ('one_hot_encoder', OneHotEncoder(drop='first'), coluna_one_hot_encoder),
        ('standard_scaler', StandardScaler(), colunas_standard_scaler),
        ('power_transform', PowerTransformer(), colunas_power_scaler)
    ],
)
```

```python
#Separando meu target das features

#Features
X = df.drop(columns=coluna_target)

#Target
y = df[coluna_target]
```

```python
#Verificando se meu modelo fica melhor sem transforma√ß√£o de target, com power ou
#quantile

param_grid = {
    'regressor__preprocessor': [preprocessamento_simples, preprocessamento_categoricas, preprocessamento_completo],
    'transformer' : [None, PowerTransformer(), QuantileTransformer(output_distribution='normal')]
}
```

```python
grid_search = grid_search_cv_regressor(
    regressor=LinearRegression(),
    preprocessor=preprocessamento_completo,
    target_transformer=PowerTransformer(),
    param_grid = param_grid   
)

grid_search
```

```python
grid_search.fit(X, y)
```

```python
grid_search.best_params_
```

```python
grid_search.best_estimator_
```

```python
coefs = dataframe_coeficientes(
    grid_search.best_estimator_.regressor_['reg'].coef_,
    grid_search.best_estimator_.regressor_['preprocessor'].get_feature_names_out(),
)

coefs
```

```python
plot_coeficientes(coefs)
```

```python
#Comparando os modelos j√° treinados com meu gridsearch

regressors = {
    #Dummyregressor
    'DummyRegressor' : {
        'preprocessor': None,
        'regressor': DummyRegressor(strategy='mean'),
        'target_transformer': None,
    },

    #LinearRegressor sem transforma√ß√£o de target e s√≥ com categ√≥rcas
     'LinearRegression' : {
        'preprocessor': preprocessamento_categoricas,
        'regressor': LinearRegression(),
        'target_transformer': None,
    },

    
    #LinearRegressor com transforma√ß√£o de target quantile transformer (best param do grid search)
     'LinearRegression_grid_seach' : {
        'preprocessor': grid_search.best_estimator_.regressor_['preprocessor'],
        'regressor': grid_search.best_estimator_.regressor_['reg'],
        'target_transformer': grid_search.best_estimator_.transformer_,
    },
}

resultados = {
    nome_modelo: treinar_e_validar_modelo_regressao(X, y, **regressor)
    for nome_modelo, regressor in regressors.items()
}

df_resultados = organiza_resultados(resultados)

df_resultados
```
